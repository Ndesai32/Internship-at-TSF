**Importing Required Libraries:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import missingno as ms

from sklearn.cluster import KMeans
```

```python
df.drop('Id',axis=1,inplace=True)
df.head()
```

# Data Distribution using pairplot:

```python
plt.figure(figsize=[8,8])
sns.pairplot(data = df,hue = 'Species')
plt.show()
```

**From the above graph looks like Iris-Setosa makes a perfect cluster in all graphs**

```python
# Split data into X & y 

X = df.drop('Species',axis = 1).values
y = df['Species']
```

**Now,We are predicting Optimum value for k using Elbow Method.**

# Elbow Method:

**This is probably the most well-known method for determining the optimal number of clusters.**

```python
sum_of_squared_dist = []

for k in range(1,15):
    kmeans = KMeans(n_clusters = k,n_init=10,max_iter=300)
    kmeans = kmeans.fit(X)
    sum_of_squared_dist.append(kmeans.inertia_)  
```

```python
plt.figure(figsize=[10,8])
plt.plot(range(1,15), sum_of_squared_dist, 'bx-')
plt.xlabel('\nNo. of Clusters',size = 18,color = 'magenta')
plt.ylabel('Sum_of_squared_distances',size=18,color='magenta')
plt.title('\nElbow Method For Optimal K',size = 25,color = 'r')
plt.grid(True)
plt.show()
```

**As expected, the plot looks like an arm with a clear elbow at k = 3.**

# Applying K-Means Clustering:

```python
kmeans = KMeans(n_clusters = 3,init = 'random',n_init=10,max_iter=300)
y_pred = kmeans.fit_predict(X)
y_pred
```

# Visualise Clusters:

```python
#Plot 3-Clusters

plt.figure(figsize = [15,8])
plt.scatter(X[y_pred == 0,0],X[y_pred == 0,1],s=50,c = 'blue',label = 'Setosa')
plt.scatter(X[y_pred == 1,0],X[y_pred == 1,1],s=50,c = 'magenta',label = 'Versicolor')
plt.scatter(X[y_pred == 2,0],X[y_pred == 2,1],s=50,c = 'red',label = 'Virginica')

# Plot the Centroid

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s = 50,c = 'black',label = 'Centroid')

plt.legend(fontsize = 'x-large',fancybox = True)
plt.grid()
plt.show()
```

**So,here we have predicted Optimum value of cluster which is k=3 and we visualised in a Graph.**
